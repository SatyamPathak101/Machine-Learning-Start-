```python
import pandas as pd
```

```python
#pandas.pivot_table()
zipcode_condition_sale2 = pd.pivot_table(Raw_Housing_Data, index=["Condition of the House", "Zipcode"], values=["Sale Price"], aggfunc=np.mean)
zipcode_condition_sale2
```

```python
zipcode_condition_sale3 = pd.pivot_table(Raw_Housing_Data, index=["Zipcode"],columns=["Condition of the House"], values = ['Sale Price'], aggfunc=np.mean)
zipcode_condition_sale3
```

```python
# Benefits of Pivot Table
# *Shorter and more comprehensible code
# *Cleaner representation
```

```python
#pandas.map()
Raw_Housing_Data['Condition of the House'] = Raw_Housing_Data['Condition of the House'].map({'Good':'1','Excellent':'3','Bad':'0','Fair':'1','Okay':'0'})
Raw_Housing_Data['Condition of the House'].unique()
```

```python
#example
def year(value):
    return value.split()[-1]

Raw_Housing_Data['year_sold']= Raw_Housing_Data['Date House was Sold'].map(year)
Raw_Housing_Data['year_sold'].head()
```

```python
#Syntax of MAP 
#data["column to operate on"].map(modifier_function)

#Limitation of Map function
#it can perform an operation only on a single column at a time

#to manipulate or transform data using multiple columns, we use pandas.apply() function.
```

```python
#pandas.apply() (doing this with the bruteforce method will take upto 7-8 mins to finish)
#hundred times faster than simple python loop
def luxury_home(row):
  count = 0
  if row[0] =='Yes':
    count = count+1
  if row[1] in ['Good', 'Excellent']:
    count = count+1
  if row[2] >= 8:
    count = count+1  
  if count >= 2:
    return 'Yes'
  else:
    return "No"
  
Raw_Housing_Data['luxury_home'] = Raw_Housing_Data[['Waterfront View','Condition of the House','Overall Grade']].apply(luxury_home, axis = 1)

Raw_Housing_Data['luxury_home'].unique()
```
